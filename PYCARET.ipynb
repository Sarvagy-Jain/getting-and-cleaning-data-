```python
from pycaret.classification import *
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/AnjulaMehto/Sampling_Assignment/main/Creditcard_data.csv'
df = pd.read_csv(url,index_col=0, on_bad_lines='skip')

# print(df.head(5))
```

```python
data = df
```

```python
x = data.drop(['Class'],axis=1)
y = data['Class']
```

```python

from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)# fit predictor and target variable
x_ros, y_ros = ros.fit_resample(x, y)
y_ros.value_counts()
```

```python
Sampled_data = pd.concat([x_ros,y_ros],axis =1)
Sampled_data.shape
```

```python
exp_clf = setup(data=Sampled_data, target = 'Class', fold_shuffle=True,session_id = 123,imputation_type="iterative",categorical_iterative_imputer="lightgbm", numeric_iterative_imputer="lightgbm") 
```

```python
exp_clf.compare_models(n_select=5)
```
